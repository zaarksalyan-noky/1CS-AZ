```python
import heapq
from collections import Counter

# Класс узла необходим для построения дерева
class Node:
    def __init__(self, char, freq):
        self.char = char
        self.freq = freq
        self.left = None
        self.right = None

# Переопределяем метод для корректной работы мин-кучи
    def __lt__(self, other):
        return self.freq < other.freq

def build_huffman_tree(text: str) -> Node:
    # Подсчитываем частоту появления каждого символа
    frequency = Counter(text)
    # Создаем кучу из листовых узлов
    heap = [Node(char, freq) for char, freq in frequency.items()]
    heapq.heapify(heap)
    
    # Объединяем узлы до тех пор, пока не останется один корень
    while len(heap) > 1:
        left = heapq.heappop(heap)
        right = heapq.heappop(heap)

        # Создаем внутренний узел без символа, его частота равна сумме частот потомков
        merged = Node(None, left.freq + right.freq)
        merged.left = left
        merged.right = right
        heapq.heappush(heap, merged)
        
    return heap[0] if heap else None

def generate_codes(node: Node, current_code: str, codes: dict):
    if node is None:
        return
    
    # Если достигли листового узла, сохраняем сгенерированный бинарный код
    if node.char is not None:
        codes[node.char] = current_code
        return
    # При движении влево добавляем '0', вправо добавляем '1'
    generate_codes(node.left, current_code + "0", codes)
    generate_codes(node.right, current_code + "1", codes)

def huffman_encode(text: str) -> tuple[str, Node]:
    if not text:
        return "", None
        
    root = build_huffman_tree(text)
    codes = {}
    generate_codes(root, "", codes)
    
    #Для частного случая: текст состоит из одинаковых символов
    if len(codes) == 1:
        char = list(codes.keys())[0]
        codes[char] = "0"
    
    # Заменяем символы на их коды
    encoded_text = "".join(codes[char] for char in text)
    return encoded_text, root

# Пример использования первого блока, расскомментируйте данную часть если необходимо запустить блоки по отдельности и закомментируйте последний блок
#if __name__ == "__main__":
#    original_text = input("Введите текст для кодирования: ")
#    encoded_data, tree_root = huffman_encode(original_text)
#    print(f"Закодированный текст: {encoded_data}")
# 
```

```python
# Класс узла дублируется для корректной работы с переданным деревом
class Node:
    def __init__(self, char, freq):
        self.char = char
        self.freq = freq
        self.left = None
        self.right = None

    def __lt__(self, other):
        return self.freq < other.freq

def huffman_decode(encoded_text: str, root: Node) -> str:
    if not encoded_text or root is None:
        return ""
        
    decoded_text = []
    current_node = root
    
    # Продолжение частного случая
    if current_node.left is None and current_node.right is None:
        return current_node.char * len(encoded_text)
        
    for bit in encoded_text:
        # Спускаемся по дереву в зависимости от бита
        if bit == '0':
            current_node = current_node.left
        else:
            current_node = current_node.right
            
        # Если дошли до листа, добавляем символ в результат и возвращаемся в корень
        if current_node.char is not None:
            decoded_text.append(current_node.char)
            current_node = root
            
    return "".join(decoded_text)

# Пример использования второго блока (предполагается, что данные получены из Блока 1)
# раскомментируйте код ниже, если нужно запустить блоки по отдельности без последнего блока
#
# if __name__ == "__main__":
#     # Передаем данные, полученные на этапе кодирования
#     decoded_result = huffman_decode(encoded_data, tree_root)
#     print(f"Декодированный текст: {decoded_result}")
```
