```python
class Solution:
    def twoSum(self, nums: list[int], target: int) -> list[int]:
        # Хэш-таблица для хранения пары {значение: индекс}
        num_map = {}
        
        for i, num in enumerate(nums):
            # Вычисляем искомое второе число
            complement = target - num
            
            # Проверяем, есть ли уже искомое число в хэш-таблице
            if complement in num_map:
                # Если да, возвращаем индексы
                return [num_map[complement], i]
            
            # Если нет, добавляем текущее число и его индекс в таблицу
            num_map[num] = i
            
        return []
```

```python
class Solution:
    def containsDuplicate(self, nums: list[int]) -> bool:
        # Инициализируем пустое хэш-множество
        seen = set()
        
        for num in nums:
            # Если число уже встречалось, значит есть дубликат
            if num in seen:
                return True
            # Добавляем число в множество встреченных
            seen.add(num)
            
        # Если цикл завершился, дубликатов нет
        return False
```

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        # Анаграммы должны быть одинаковой длины
        if len(s) != len(t):
            return False
            
        # Хэш-таблица для подсчета частоты символов
        char_count = {}
        
        # Считаем символы в первой строке
        for char in s:
            char_count[char] = char_count.get(char, 0) + 1
            
        # Вычитаем частоты на основе второй строки
        for char in t:
            if char not in char_count or char_count[char] == 0:
                # Символа нет или его количество превышено
                return False
            char_count[char] -= 1
            
        return True
```

```python
class Solution:
    def intersection(self, nums1: list[int], nums2: list[int]) -> list[int]:
        # Преобразуем первый массив в хэш-множество для быстрого поиска
        set1 = set(nums1)
        # Множество для хранения уникальных элементов пересечения
        intersection_set = set()
        
        # Проверяем элементы второго массива на наличие в первом множестве
        for num in nums2:
            if num in set1:
                intersection_set.add(num)
                
        # Конвертируем множество обратно в список для ответа
        return list(intersection_set)
```

```python
from collections import defaultdict

class Solution:
    def groupAnagrams(self, strs: list[str]) -> list[list[str]]:
        # Используем defaultdict, чтобы автоматически создавать пустой список 
        # для новых ключей (сигнатур)
        anagram_map = defaultdict(list)
        
        for s in strs:
            # Сортируем строку, чтобы получить унифицированный ключ для анаграмм
            # tuple(sorted(s)) создает неизменяемый ключ, который может быть в хэш-таблице
            sorted_key = tuple(sorted(s))
            
            # Добавляем исходное слово в соответствующую группу
            anagram_map[sorted_key].append(s)
            
        # Возвращаем список всех сгруппированных значений
        return list(anagram_map.values())
```
