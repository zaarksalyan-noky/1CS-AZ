```python
class Solution:
    def isValid(self, s: str) -> bool:
        # Словарь для быстрого поиска соответствующих пар скобок
        mapping = {")": "(", "}": "{", "]": "["}
        # Инициализируем стек в виде списка
        stack = []
        
        for char in s:
            # Если символ является закрывающей скобкой
            if char in mapping:
                # Извлекаем верхний элемент, если стек не пуст, иначе подставляем фиктивное значение
                top_element = stack.pop() if stack else '#'
                # Если скобки не совпадают, строка невалидна
                if mapping[char] != top_element:
                    return False
            else:
                # Если символ - открывающая скобка, кладем в стек
                stack.append(char)
                
        # Строка валидна только в том случае, если стек остался пустым
        return not stack
```

```python
class MinStack:
    def __init__(self):
        # Основной стек для хранения элементов
        self.stack = []
        # Вспомогательный стек для хранения минимальных значений
        self.min_stack = []

    def push(self, val: int) -> None:
        self.stack.append(val)
        # Если min_stack пуст или новое значение меньше либо равно текущему минимуму, добавляем его
        if not self.min_stack or val <= self.min_stack[-1]:
            self.min_stack.append(val)

    def pop(self) -> None:
        # Если удаляемый элемент является текущим минимумом, удаляем его и из min_stack
        if self.stack[-1] == self.min_stack[-1]:
            self.min_stack.pop()
        self.stack.pop()

    def top(self) -> int:
        return self.stack[-1]

    def getMin(self) -> int:
        # Вершина min_stack всегда содержит текущий минимум
        return self.min_stack[-1]
```

```python
class Solution:
    def evalRPN(self, tokens: list[str]) -> int:
        stack = []
        
        for token in tokens:
            if token not in {"+", "-", "*", "/"}:
                # Если токен - число, переводим в int и кладем в стек
                stack.append(int(token))
            else:
                # Извлекаем два последних числа (порядок важен для вычитания и деления)
                b = stack.pop()
                a = stack.pop()
                
                # Выполняем операцию и кладем результат обратно
                if token == "+":
                    stack.append(a + b)
                elif token == "-":
                    stack.append(a - b)
                elif token == "*":
                    stack.append(a * b)
                elif token == "/":
                    # В Python обычное деление с int() усекает дробную часть к нулю, как требует условие
                    stack.append(int(a / b))
                    
        # В конце в стеке останется ровно один элемент — ответ
        return stack[0]
```

```python
class MyQueue:
    def __init__(self):
        # Стек для входа
        self.in_stack = []
        # Стек для выхода (развернутые элементы)
        self.out_stack = []

    def push(self, x: int) -> None:
        # Всегда добавляем во входной стек
        self.in_stack.append(x)

    def pop(self) -> int:
        self.peek() # Гарантируем, что out_stack не пуст
        return self.out_stack.pop()

    def peek(self) -> int:
        # Если выходной стек пуст, перекладываем в него все из входного
        if not self.out_stack:
            while self.in_stack:
                self.out_stack.append(self.in_stack.pop())
        return self.out_stack[-1]

    def empty(self) -> bool:
        # Очередь пуста, если оба стека пусты
        return not self.in_stack and not self.out_stack
```

```python
class MyCircularQueue:
    def __init__(self, k: int):
        self.queue = [0] * k # Фиксированный массив
        self.capacity = k
        self.count = 0
        self.headIndex = 0

    def enQueue(self, value: int) -> bool:
        if self.isFull():
            return False
        # Вычисляем индекс для вставки нового элемента по кругу
        tailIndex = (self.headIndex + self.count) % self.capacity
        self.queue[tailIndex] = value
        self.count += 1
        return True

    def deQueue(self) -> bool:
        if self.isEmpty():
            return False
        # Сдвигаем указатель головы по кругу
        self.headIndex = (self.headIndex + 1) % self.capacity
        self.count -= 1
        return True

    def Front(self) -> int:
        if self.isEmpty():
            return -1
        return self.queue[self.headIndex]

    def Rear(self) -> int:
        if self.isEmpty():
            return -1
        tailIndex = (self.headIndex + self.count - 1) % self.capacity
        return self.queue[tailIndex]

    def isEmpty(self) -> bool:
        return self.count == 0

    def isFull(self) -> bool:
        return self.count == self.capacity
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class MyLinkedList:
    def __init__(self):
        self.size = 0
        # Использование dummy узла избавляет от множества краевых случаев
        self.head = ListNode(0)

    def get(self, index: int) -> int:
        if index < 0 or index >= self.size:
            return -1
        
        curr = self.head.next
        for _ in range(index):
            curr = curr.next
        return curr.val

    def addAtHead(self, val: int) -> None:
        self.addAtIndex(0, val)

    def addAtTail(self, val: int) -> None:
        self.addAtIndex(self.size, val)

    def addAtIndex(self, index: int, val: int) -> None:
        if index > self.size:
            return
        if index < 0:
            index = 0
            
        self.size += 1
        pred = self.head
        # Находим узел-предшественник
        for _ in range(index):
            pred = pred.next
            
        # Вставляем новый узел
        new_node = ListNode(val, pred.next)
        pred.next = new_node

    def deleteAtIndex(self, index: int) -> None:
        if index < 0 or index >= self.size:
            return
            
        self.size -= 1
        pred = self.head
        # Находим узел-предшественник
        for _ in range(index):
            pred = pred.next
            
        # Пропускаем удаляемый узел
        pred.next = pred.next.next
```
